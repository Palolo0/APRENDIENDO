```python
#Este programa compara dos archivos celda por celda para ver las diferencias.
import csv
import pandas as pd

def comparar_celdas(archivo1, archivo2, archivo_salida):
    diferencias = []  # Lista para almacenar las diferencias

    with open(archivo1, 'r',encoding='utf-8') as t1, open(archivo2, 'r',encoding='utf-8') as t2:
        lector1 = csv.reader(t1)
        lector2 = csv.reader(t2)

        for fila1, fila2 in zip(lector1, lector2):
            for celda1, celda2 in zip(fila1, fila2):
                if celda1 != celda2:
                    diferencias.append([celda1, celda2])

    # Crear un DataFrame con las diferencias
    df_diferencias = pd.DataFrame(diferencias, columns=['Archivo 1', 'Archivo 2'])

    # Exportar el DataFrame a un archivo CSV
    df_diferencias.to_csv(archivo_salida, index=False)

# Llama a la función con los nombres de tus archivos CSV y el nombre del archivo de salida
comparar_celdas('archivo_grande_bien.csv', 'archivo_grande_mal.csv', 'archivo_diferencias.csv')

```
